### 2.1 Data, descriptive statistics, regimes, covariance and bootstrap

```python
def _max_drawdown_from_returns(r: pd.Series) -> float:
    """Return maximum drawdown from a daily simple-return series."""
    wealth = (1.0 + r.dropna()).cumprod()
    dd = wealth / wealth.cummax() - 1.0
    return float(dd.min())


def make_etf_summary_stats(R: pd.DataFrame, tickers=None, ann_factor: int = 252) -> pd.DataFrame:
    """
    Descriptive statistics for ETF universe.
    Returns, volatility and MaxDD are reported in percentage points.
    Sharpe assumes zero risk-free rate for descriptive purposes.
    """
    if tickers is None:
        tickers = list(R.columns)

    rows = []
    for t in tickers:
        r = R[t].dropna()
        if len(r) == 0:
            continue

        ann_ret = (1.0 + r).prod() ** (ann_factor / len(r)) - 1.0
        ann_vol = r.std(ddof=1) * np.sqrt(ann_factor)
        sharpe = ann_ret / ann_vol if ann_vol > 0 else np.nan

        rows.append({
            "Ticker": t,
            "Start": r.index.min().date(),
            "End": r.index.max().date(),
            "Obs": int(len(r)),
            "Ann. Ret": 100.0 * ann_ret,
            "Ann. Vol": 100.0 * ann_vol,
            "Sharpe": sharpe,
            "Skew": r.skew(),
            "Excess Kurt": r.kurt(),
            "MaxDD": 100.0 * _max_drawdown_from_returns(r),
        })

    out = pd.DataFrame(rows).set_index("Ticker")
    return out


def rolling_equicorr(R, W_lag, window=22):
    rho_vals, dates = [], []
    for i in range(window, len(R)):
        cov  = R.iloc[i - window:i].cov()
        date = R.index[i]
        w    = W_lag.loc[date].dropna()
        cov  = cov.loc[w.index, w.index]
        sig2 = np.diag(cov.values)
        sig  = np.sqrt(sig2)
        wv   = w.values
        port_var  = float(wv @ cov.values @ wv)
        diag_term = float((wv**2) @ sig2)
        a     = wv * sig
        denom = float(a.sum()**2 - (a**2).sum())
        rho_vals.append((port_var - diag_term) / denom if denom > 0 else np.nan)
        dates.append(date)
    return pd.Series(rho_vals, index=dates, name='equicorr').dropna()


def rolling_dcor_gap(R, W_lag, window=22):
    gap_vals, dates = [], []
    cols = list(R.columns)
    n = len(cols)

    for i in range(window, len(R)):
        Rw  = R.iloc[i - window:i]
        date = R.index[i]

        # keep alignment consistent with other indicators:
        # W_lag is already shifted(1); using date gives a 1-day lag w.r.t. prices
        w = W_lag.loc[date].reindex(cols).astype(float).values

        rho_abs = np.abs(Rw.corr().values.astype(float))

        # build symmetric dCor matrix (upper triangle only)
        dcor_mat = np.eye(n, dtype=float)
        for a in range(n):
            xa = Rw.iloc[:, a].values
            for b in range(a + 1, n):
                xb = Rw.iloc[:, b].values
                d = dcor.distance_correlation(xa, xb)
                dcor_mat[a, b] = d
                dcor_mat[b, a] = d

        gap_vals.append(float(w @ dcor_mat @ w - w @ rho_abs @ w))
        dates.append(date)

    return pd.Series(gap_vals, index=dates, name='gap')


def rolling_frobenius(R, window=22, rho_lt=None, normalise=True):
    cols = list(R.columns)
    n = len(cols)
    if rho_lt is None:
        rho_lt = R.corr().values.astype(float)

    frob_vals, dates = [], []
    scale = math.sqrt(n * (n - 1)) if (normalise and n > 1) else 1.0

    for i in range(window, len(R)):
        rho_t = R.iloc[i - window:i].corr().values.astype(float)
        frob_vals.append(np.linalg.norm(rho_t - rho_lt, ord='fro') / scale)
        dates.append(R.index[i])

    return pd.Series(frob_vals, index=dates, name='frob_d')


def to_full_index(mask, full_idx):
    s = pd.Series(False, index=full_idx)
    s.update(mask)
    return s


def compute_cov_lw(R, mask=None, label=''):
    R_sub = R.loc[mask] if mask is not None else R
    if len(R_sub) < R.shape[1] + 2:
        print(f'  Warning [{label}]: only {len(R_sub)} obs -> fallback to full sample')
        R_sub = R
    lw = LedoitWolf(assume_centered=False)
    lw.fit(R_sub.values)
    cov = pd.DataFrame(lw.covariance_, index=R.columns, columns=R.columns)
    print(f'  [{label}] N={len(R_sub)}, LW-shrinkage={lw.shrinkage_:.4f}')
    return cov


def cov_to_corr(cov_df):
    D = np.diag(1 / np.sqrt(np.diag(cov_df.values)))
    return pd.DataFrame(D @ cov_df.values @ D, index=cov_df.index, columns=cov_df.columns)


def stationary_bootstrap_means(data_np, block_size=22, n_boot=500, seed=42):
    rng  = np.random.default_rng(seed)
    T, p = data_np.shape
    boot_means = np.empty((n_boot, p))
    for b in range(n_boot):
        idx = []
        while len(idx) < T:
            start  = rng.integers(0, T)
            length = rng.geometric(1.0 / block_size)
            length = min(length, T - len(idx))
            idx.extend((start + j) % T for j in range(length))
        boot_means[b] = data_np[idx[:T]].mean(axis=0)
    return boot_means


def effective_sample_size(weights):
    """Kish effective sample size for normalized non-negative weights."""
    w = np.asarray(weights, dtype=float)
    w = w / np.sum(w)
    return float(1.0 / np.sum(w**2))


def softmax_stress_weights(signal, kappa=1.0, clip_z=3.0):
    """
    Convert a regime/stress signal into continuous observation weights.

    kappa = 0   -> equal weights (unconditional covariance)
    kappa > 0   -> increasingly stress-conditioned covariance
    """
    x = pd.Series(signal).dropna().astype(float)
    z = (x - x.mean()) / x.std(ddof=0)
    z = z.clip(-clip_z, clip_z)

    # subtract max for numerical stability
    raw = np.exp(kappa * (z - z.max()))
    w = raw / raw.sum()
    return pd.Series(w.values, index=x.index, name=f"soft_weights_kappa_{kappa:g}")


def weighted_covariance(R_sub, weights):
    """
    Weighted sample covariance using normalized observation weights.
    R_sub: DataFrame returns
    weights: Series aligned with R_sub.index
    """
    w = pd.Series(weights).reindex(R_sub.index).dropna().astype(float)
    X_df = R_sub.loc[w.index]
    X = X_df.values.astype(float)
    wv = (w.values / w.values.sum()).reshape(-1, 1)

    mu = (wv * X).sum(axis=0)
    Xc = X - mu
    cov = (wv * Xc).T @ Xc

    return pd.DataFrame(cov, index=R_sub.columns, columns=R_sub.columns)


def constant_corr_target(cov):
    """
    Constant-correlation shrinkage target preserving variances.
    """
    cov = pd.DataFrame(cov).astype(float)
    vols = np.sqrt(np.maximum(np.diag(cov.values), 1e-16))
    corr = cov.values / np.outer(vols, vols)
    np.fill_diagonal(corr, np.nan)
    rho_bar = np.nanmean(corr)

    target = rho_bar * np.outer(vols, vols)
    np.fill_diagonal(target, vols**2)

    return pd.DataFrame(target, index=cov.index, columns=cov.columns), float(rho_bar)


def shrink_cov_constant_corr(cov, delta="auto", n_eff=None):
    """
    Manual shrinkage of a covariance matrix toward a constant-correlation target.

    If delta='auto', use a conservative rule increasing shrinkage when N_eff is small:
        delta = clip(N / N_eff, 0.02, 0.50)

    This is intentionally simple for V1.1; a weighted Ledoit-Wolf estimator can
    replace it later.
    """
    cov = pd.DataFrame(cov).astype(float)
    target, rho_bar = constant_corr_target(cov)

    if delta == "auto":
        if n_eff is None or not np.isfinite(n_eff) or n_eff <= 0:
            d = 0.10
        else:
            d = float(np.clip(cov.shape[0] / n_eff, 0.02, 0.50))
    else:
        d = float(delta)

    shrunk = (1.0 - d) * cov + d * target
    return shrunk, d, rho_bar


def compute_soft_frobenius_cov(R_train, window=22, kappa=1.0,
                               clip_z=3.0, shrink_delta="auto",
                               label="soft-frob"):
    """
    Estimate a soft stress-weighted covariance matrix from Frobenius regime intensity.
    Long-run correlation is estimated on TRAIN only.
    """
    rho_lt_train = R_train.corr().values.astype(float)
    frob_train = rolling_frobenius(R_train, window=window, rho_lt=rho_lt_train, normalise=True)

    omega = softmax_stress_weights(frob_train, kappa=kappa, clip_z=clip_z)
    n_eff = effective_sample_size(omega.values)

    cov_w = weighted_covariance(R_train, omega)
    cov_shr, delta, rho_bar = shrink_cov_constant_corr(cov_w, delta=shrink_delta, n_eff=n_eff)

    print(f"  [{label}] kappa={kappa:.2f}  N_eff={n_eff:.1f}/{len(omega)}  "
          f"shrink_delta={delta:.4f}  target_rho={rho_bar:.4f}")

    return {
        "Sigma_stress": cov_shr,
        "Sigma_weighted_raw": cov_w,
        "weights": omega,
        "frob_signal": frob_train,
        "n_eff": n_eff,
        "shrink_delta": delta,
        "target_rho": rho_bar,
        "rho_lt_train": rho_lt_train,
    }


def corr_from_cov(cov):
    vals = cov.values if isinstance(cov, pd.DataFrame) else np.asarray(cov)
    d = np.sqrt(np.diag(vals))
    inv = np.diag(1 / np.where(d > 0, d, np.nan))
    corr = inv @ vals @ inv
    if isinstance(cov, pd.DataFrame):
        return pd.DataFrame(corr, index=cov.index, columns=cov.columns)
    return corr


def portfolio_risk_contributions(weights, cov):
    w = pd.Series(weights).reindex(cov.columns).fillna(0.0).values.astype(float)
    S = cov.values.astype(float)
    total = float(w @ S @ w)
    if total <= 0:
        return pd.Series(np.nan, index=cov.columns)
    mrc = S @ w
    rc = w * mrc / total
    return pd.Series(rc, index=cov.columns)


def plot_heatmap(df, title, figsize=(8, 6), cmap="coolwarm"):
    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(df.values, aspect="auto", cmap=cmap)
    ax.set_xticks(range(len(df.columns)))
    ax.set_xticklabels(df.columns, rotation=90)
    ax.set_yticks(range(len(df.index)))
    ax.set_yticklabels(df.index)
    ax.set_title(title)
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.show()
```

### 2.2 Optimization, frontiers, ablation and validation helpers

```python
def solve_socp(Sigma_stress_np, Sigma_growth_np, mu_mean_annual_np,
               selected_set, equity_idx, w_core_min, tau_annual, k=1.0):
    n   = len(mu_mean_annual_np)
    reg = 1e-10

    # Cholesky factor for growth uncertainty covariance (annual units)
    try:
        L_g = np.linalg.cholesky(Sigma_growth_np + reg * np.eye(n))
    except np.linalg.LinAlgError:
        L_g = np.diag(np.sqrt(np.maximum(np.diag(Sigma_growth_np), 0) + reg))

    Ss_reg = Sigma_stress_np + reg * np.eye(n)

    w = cp.Variable(n)
    constraints = [
        cp.sum(w) == 1,
        w >= 0,
        w[equity_idx] >= w_core_min,
        mu_mean_annual_np @ w - k * cp.norm(L_g @ w, 2) >= tau_annual,
    ]

    excluded = [i for i in range(n) if i not in selected_set]
    if excluded:
        constraints.append(w[np.array(excluded)] == 0)

    prob = cp.Problem(cp.Minimize(cp.quad_form(w, Ss_reg)), constraints)

    # Prefer ECOS for SOCP; fallback to SCS if needed.
    try:
        prob.solve(solver=cp.ECOS, verbose=False)
    except Exception:
        try:
            prob.solve(solver=cp.SCS, verbose=False)
        except Exception:
            return None, np.inf

    if prob.status in ('optimal', 'optimal_inaccurate') and w.value is not None:
        return w.value.copy(), float(prob.value)
    return None, np.inf


def greedy_core_satellite(
    tickers, Sigma_stress, Sigma_growth, mu_mean_annual,
    equity_ticker=EQUITY, w_core_min=0.50, tau=0.0, k=1.0, verbose=True,
):
    n      = len(tickers)
    eq_idx = tickers.index(equity_ticker)
    sat_ix = [i for i in range(n) if i != eq_idx]

    Ss = Sigma_stress.loc[tickers, tickers].values
    Sg = Sigma_growth.loc[tickers, tickers].values
    mu = mu_mean_annual[tickers].values

    selected = {eq_idx}
    w_cur    = np.zeros(n); w_cur[eq_idx] = 1.0
    risk_cur = float(w_cur @ Ss @ w_cur)

    for rnd in range(len(sat_ix)):
        best_risk, best_w, best_add = risk_cur, None, None
        for s in sat_ix:
            if s in selected:
                continue
            w_opt, obj = solve_socp(Ss, Sg, mu, selected | {s}, eq_idx, w_core_min, tau, k)
            if w_opt is not None and obj < best_risk - 1e-10:
                best_risk, best_w, best_add = obj, w_opt, s

        if best_add is not None:
            selected |= {best_add}
            risk_cur  = best_risk
            w_cur     = best_w
            if verbose:
                ann_sv = np.sqrt(max(risk_cur * 252, 0))
                print(f'  Round {rnd+1}: +{tickers[best_add]:<10} stress_var={risk_cur:.6f}  (ann. stress_vol={ann_sv:.4f})')
        else:
            if verbose:
                print(f'  Round {rnd+1}: no further improvement -> stop')
            break

    w_s = pd.Series(np.maximum(w_cur, 0), index=tickers)
    w_s /= w_s.sum()

    # Report: stress vol (annual), mean growth (annual), robust growth (annual)
    stress_vol = np.sqrt(max(float(w_s.values @ Ss @ w_s.values) * 252, 0))
    mean_g     = float(mu @ w_s.values)
    robust_g   = float(mean_g - k * np.sqrt(max(w_s.values @ Sg @ w_s.values, 0)))

    return {
        'weights': w_s,
        'selected': [tickers[i] for i in sorted(selected)],
        'stress_vol_annual': stress_vol,
        'mean_growth_annual': mean_g,
        'robust_growth_annual': robust_g,
        'tau': tau,
        'w_core_min': w_core_min,
        'k': k,
    }


def compute_indicator_series(R_sub, indicator='equicorr', window=22, W_lag_sub=None, rho_lt_train=None):
    if indicator == 'equicorr':
        return rolling_equicorr(R_sub, W_lag_sub, window)
    elif indicator == 'gap':
        return rolling_dcor_gap(R_sub, W_lag_sub, window)
    elif indicator == 'frob':
        return rolling_frobenius(R_sub, window, rho_lt=rho_lt_train, normalise=True)
    else:
        raise ValueError(f'Unknown indicator: {indicator}')


def estimate_train_objects(R_train, logR_train, indicator='equicorr',
                           window=22, percentile=80, block_size=22, n_boot=1000):
    # Proxy weights aligned to the train sample
    W_tr_lag = W_lag.loc[R_train.index]

    # Long-run correlation for Frobenius must be estimated on TRAIN only
    rho_lt_train = R_train.corr().values.astype(float)

    ind_series = compute_indicator_series(
        R_train, indicator=indicator, window=window, W_lag_sub=W_tr_lag, rho_lt_train=rho_lt_train
    )
    stress_mask = to_full_index(ind_series > np.percentile(ind_series, percentile), R_train.index)

    Sigma_stress_tr = compute_cov_lw(R_train, stress_mask, f'train-{indicator}-stress')
    Sigma_base_tr   = compute_cov_lw(R_train, label=f'train-{indicator}-full')

    boot_tr = stationary_bootstrap_means(logR_train.values, block_size, n_boot)
    mu_mean_tr = pd.Series(boot_tr.mean(0) * 252, index=R_train.columns)
    Sigma_growth_tr = pd.DataFrame(np.cov((boot_tr * 252), rowvar=False),
                                   index=R_train.columns, columns=R_train.columns)

    return {
        'indicator_series': ind_series,
        'stress_mask': stress_mask,
        'Sigma_stress': Sigma_stress_tr,
        'Sigma_base': Sigma_base_tr,
        'mu_mean': mu_mean_tr,
        'Sigma_growth': Sigma_growth_tr,
        'rho_lt_train': rho_lt_train,
    }


def perf(ret, name):
    r    = ret.dropna()
    cum  = (1 + r).cumprod()
    ann_r = cum.iloc[-1] ** (252 / len(r)) - 1
    ann_v = r.std() * np.sqrt(252)
    sh    = ann_r / ann_v if ann_v > 0 else 0
    dd    = (cum / cum.expanding().max() - 1).min()
    calmar = ann_r / abs(dd) if dd != 0 else np.nan
    print(f'  {name:<32} Ret={ann_r:+.3f}  Vol={ann_v:.3f}  '
          f'Sharpe={sh:.3f}  MaxDD={dd:.3f}  Calmar={calmar:.3f}')
    return dict(name=name, ret=ann_r, vol=ann_v, sharpe=sh, maxdd=dd, calmar=calmar)


def perf_summary(ret_s):
    ann_ret = float(np.exp(np.log1p(ret_s).mean() * 252) - 1)
    ann_vol = float(ret_s.std() * np.sqrt(252))
    sharpe  = ann_ret / ann_vol if ann_vol > 0 else np.nan
    cum     = (1 + ret_s).cumprod()
    dd      = cum / cum.cummax() - 1
    maxdd   = float(dd.min())
    calmar  = ann_ret / abs(maxdd) if maxdd < 0 else np.nan
    return {'ret': ann_ret, 'vol': ann_vol, 'sharpe': sharpe, 'maxdd': maxdd, 'calmar': calmar}


def estimate_train_objects_soft_frob(R_train, logR_train,
                                     window=22, kappa=1.5, clip_z=3.0,
                                     shrink_delta="auto",
                                     block_size=22, n_boot=1000):
    """
    Train-only estimator for the V1.1 soft Frobenius specification:
    - soft stress-weighted covariance
    - stationary-bootstrap growth mean/covariance
    - full-sample LW covariance for Markowitz baseline
    """
    soft = compute_soft_frobenius_cov(
        R_train, window=window, kappa=kappa, clip_z=clip_z,
        shrink_delta=shrink_delta, label="train-soft-frob"
    )

    Sigma_base_tr = compute_cov_lw(R_train, label="train-full-lw-for-markowitz")

    boot_tr = stationary_bootstrap_means(logR_train.values, block_size, n_boot)
    mu_mean_tr = pd.Series(boot_tr.mean(0) * 252, index=R_train.columns)
    Sigma_growth_tr = pd.DataFrame(
        np.cov(boot_tr * 252, rowvar=False),
        index=R_train.columns, columns=R_train.columns
    )

    soft.update({
        "Sigma_base": Sigma_base_tr,
        "Sigma_growth": Sigma_growth_tr,
        "mu_mean": mu_mean_tr,
        "boot_means": boot_tr,
        "kappa": kappa,
    })
    return soft


def solve_markowitz_socp(Sigma_cov, Sigma_growth, mu_mean_annual,
                         tau_annual, k=1.0, w_core_min=0.50,
                         equity_ticker=EQUITY, tickers=TICKERS):
    """
    Markowitz-style efficient frontier point:
    minimize w' Sigma_cov w subject to robust growth >= tau,
    fully invested, long-only, and equity core minimum.

    No greedy/sparsity: all assets are investible.
    """
    n = len(tickers)
    eq_idx = tickers.index(equity_ticker)

    S = Sigma_cov.loc[tickers, tickers].values
    Sg = Sigma_growth.loc[tickers, tickers].values
    mu = mu_mean_annual[tickers].values

    selected_all = set(range(n))
    w_opt, obj = solve_socp(S, Sg, mu, selected_all, eq_idx,
                            w_core_min, tau_annual, k=k)
    if w_opt is None:
        w_s = pd.Series(np.nan, index=tickers)
        return {
            "tau": tau_annual, "weights": w_s, "objective": np.inf,
            "stress_vol_annual": np.nan, "mean_growth_annual": np.nan,
            "robust_growth_annual": np.nan,
        }

    w_s = pd.Series(np.maximum(w_opt, 0), index=tickers)
    w_s /= w_s.sum()

    # For Markowitz, objective risk is the covariance passed to the solver.
    obj_vol = float(np.sqrt(max(w_s.values @ S @ w_s.values * 252, 0)))
    try:
        L_g = np.linalg.cholesky(Sg + 1e-10*np.eye(n))
    except np.linalg.LinAlgError:
        L_g = np.diag(np.sqrt(np.maximum(np.diag(Sg), 0) + 1e-10))

    mean_g = float(mu @ w_s.values)
    rob_g  = float(mean_g - k * np.linalg.norm(L_g @ w_s.values))

    return {
        "tau": tau_annual,
        "weights": w_s,
        "objective": float(obj),
        "objective_vol_annual": obj_vol,
        "mean_growth_annual": mean_g,
        "robust_growth_annual": rob_g,
        "selected": list(w_s[w_s > 1e-6].index),
    }


def portfolio_returns_static(R_oos, weights):
    w = pd.Series(weights).reindex(R_oos.columns).fillna(0.0).values
    return pd.Series(R_oos.values @ w, index=R_oos.index)


def extended_perf_summary(ret_s, rf=0.0):
    """
    OOS metrics for a daily return series.
    Martin Ratio = (CAGR - rf) / Ulcer Index.
    """
    r = ret_s.dropna().astype(float)
    if len(r) == 0:
        return {k: np.nan for k in ["ret","vol","sharpe","maxdd","calmar","ulcer","martin"]}

    ann_ret = float(np.exp(np.log1p(r).mean() * 252) - 1)
    ann_vol = float(r.std() * np.sqrt(252))
    sharpe  = float((ann_ret - rf) / ann_vol) if ann_vol > 0 else np.nan

    cum = (1 + r).cumprod()
    dd  = cum / cum.cummax() - 1
    maxdd = float(dd.min())
    calmar = float(ann_ret / abs(maxdd)) if maxdd < 0 else np.nan

    ulcer = float(np.sqrt(np.mean(np.square(np.minimum(dd.values, 0)))))
    martin = float((ann_ret - rf) / ulcer) if ulcer > 0 else np.nan

    return {
        "ret": ann_ret,
        "vol": ann_vol,
        "sharpe": sharpe,
        "maxdd": maxdd,
        "calmar": calmar,
        "ulcer": ulcer,
        "martin": martin,
    }


def build_frontier_comparison(train_obj, tau_grid, R_test,
                              w_core_min=0.50, k=1.0,
                              tickers=TICKERS):
    """
    Build comparable OOS frontiers:
    - Greedy SAA uses soft-Frobenius stress covariance and sparse outer loop.
    - Markowitz uses full-sample LW covariance and all assets.
    Both use the same robust-growth constraint and same tau grid.
    """
    rows = []
    weights_greedy = {}
    weights_markowitz = {}

    for tau in tau_grid:
        # Greedy SAA frontier point
        g_res = greedy_core_satellite(
            tickers,
            train_obj["Sigma_stress"],
            train_obj["Sigma_growth"],
            train_obj["mu_mean"],
            w_core_min=w_core_min,
            tau=float(tau),
            k=k,
            verbose=False
        )
        g_ret = portfolio_returns_static(R_test, g_res["weights"])
        g_perf = extended_perf_summary(g_ret)
        g_perf.update({
            "model": "Greedy Soft-Frob SAA",
            "tau": float(tau),
            "n_assets": int((g_res["weights"] > 1e-6).sum()),
            "train_obj_vol": g_res["stress_vol_annual"],
            "train_meanG": g_res["mean_growth_annual"],
            "train_robustG": g_res["robust_growth_annual"],
        })
        rows.append(g_perf)
        weights_greedy[float(tau)] = g_res["weights"]

        # Markowitz robust-growth frontier point
        m_res = solve_markowitz_socp(
            train_obj["Sigma_base"],
            train_obj["Sigma_growth"],
            train_obj["mu_mean"],
            tau_annual=float(tau),
            k=k,
            w_core_min=w_core_min,
            tickers=tickers
        )
        if m_res["weights"].isna().any():
            m_ret = pd.Series(np.nan, index=R_test.index)
            m_perf = extended_perf_summary(m_ret)
        else:
            m_ret = portfolio_returns_static(R_test, m_res["weights"])
            m_perf = extended_perf_summary(m_ret)

        m_perf.update({
            "model": "Markowitz Robust-Growth",
            "tau": float(tau),
            "n_assets": int((m_res["weights"] > 1e-6).sum()) if not m_res["weights"].isna().any() else 0,
            "train_obj_vol": m_res.get("objective_vol_annual", np.nan),
            "train_meanG": m_res["mean_growth_annual"],
            "train_robustG": m_res["robust_growth_annual"],
        })
        rows.append(m_perf)
        weights_markowitz[float(tau)] = m_res["weights"]

    frontier_tbl = pd.DataFrame(rows)
    return frontier_tbl, weights_greedy, weights_markowitz


def _weight_flags(weights, equity_ticker=EQUITY, eps=1e-6):
    """Basic flags to separate genuine core-satellite points from SPY-only endpoints."""
    w = pd.Series(weights).astype(float)
    if w.isna().any():
        return {
            "n_assets": 0,
            "equity_weight": np.nan,
            "is_spy_only": False,
            "is_core_satellite": False,
        }

    n_assets = int((w.abs() > eps).sum())
    eq_w = float(w.get(equity_ticker, 0.0))
    is_spy_only = bool(eq_w >= 1.0 - 1e-4 and n_assets <= 1)
    return {
        "n_assets": n_assets,
        "equity_weight": eq_w,
        "is_spy_only": is_spy_only,
        "is_core_satellite": bool((not is_spy_only) and n_assets > 1),
    }


def solve_greedy_point(Sigma_cov, Sigma_growth, mu_mean, tau,
                       w_core_min=W_CORE_MIN, k=K_SOCP, tickers=TICKERS,
                       model_name="Greedy"):
    """Greedy frontier point with arbitrary covariance in the objective."""
    return greedy_core_satellite(
        tickers, Sigma_cov, Sigma_growth, mu_mean,
        w_core_min=w_core_min, tau=float(tau), k=k, verbose=False
    )


def solve_all_assets_point(Sigma_cov, Sigma_growth, mu_mean, tau,
                           w_core_min=W_CORE_MIN, k=K_SOCP, tickers=TICKERS,
                           model_name="All-assets MV"):
    """All-assets Markowitz/SOCP frontier point with arbitrary covariance in the objective."""
    return solve_markowitz_socp(
        Sigma_cov, Sigma_growth, mu_mean,
        tau_annual=float(tau), k=k, w_core_min=w_core_min, tickers=tickers
    )


def build_ablation_frontiers(train_obj, tau_grid, R_eval,
                             w_core_min=W_CORE_MIN, k=K_SOCP,
                             tickers=TICKERS):
    """
    Build four comparable frontiers:
      1) Markowitz Base-Cov
      2) Stress-MV Soft-Frob
      3) Greedy Base-Cov
      4) Greedy Soft-Frob

    All use the same robust-growth constraint and same tau grid.
    """
    model_specs = [
        ("Markowitz Base-Cov", "all",    train_obj["Sigma_base"]),
        ("Stress-MV Soft-Frob", "all",  train_obj["Sigma_stress"]),
        ("Greedy Base-Cov", "greedy",   train_obj["Sigma_base"]),
        ("Greedy Soft-Frob", "greedy",  train_obj["Sigma_stress"]),
    ]

    rows = []
    weights_by_model_tau = {}

    for tau in tau_grid:
        for model_name, solver_type, Sigma_obj in model_specs:
            if solver_type == "greedy":
                res = solve_greedy_point(
                    Sigma_obj, train_obj["Sigma_growth"], train_obj["mu_mean"],
                    tau=tau, w_core_min=w_core_min, k=k, tickers=tickers,
                    model_name=model_name
                )
            else:
                res = solve_all_assets_point(
                    Sigma_obj, train_obj["Sigma_growth"], train_obj["mu_mean"],
                    tau=tau, w_core_min=w_core_min, k=k, tickers=tickers,
                    model_name=model_name
                )

            w = res["weights"]
            weights_by_model_tau[(model_name, float(tau))] = w

            if w.isna().any():
                perf_row = extended_perf_summary(pd.Series(np.nan, index=R_eval.index))
                flags = _weight_flags(w)
            else:
                ret_eval = portfolio_returns_static(R_eval, w)
                perf_row = extended_perf_summary(ret_eval)
                flags = _weight_flags(w)

            perf_row.update({
                "model": model_name,
                "tau": float(tau),
                "train_obj_vol": res.get("stress_vol_annual", res.get("objective_vol_annual", np.nan)),
                "train_meanG": res.get("mean_growth_annual", np.nan),
                "train_robustG": res.get("robust_growth_annual", np.nan),
                **flags,
            })
            rows.append(perf_row)

    out = pd.DataFrame(rows)
    return out, weights_by_model_tau


def plot_ablation_frontiers(ablation_tbl, exclude_spy_only=False, title_suffix=""):
    dfp = ablation_tbl.copy()
    if exclude_spy_only:
        dfp = dfp[~dfp["is_spy_only"]].copy()

    frontier_wide = {
        model: df.set_index("tau").sort_index()
        for model, df in dfp.groupby("model")
    }

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    ax = axes[0, 0]
    for model, dfm in frontier_wide.items():
        ax.plot(dfm["vol"], dfm["ret"], marker="o", label=model)
    ax.set_title(f"CAGR vs Volatility {title_suffix}")
    ax.set_xlabel("Volatility")
    ax.set_ylabel("CAGR")
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=8)

    ax = axes[0, 1]
    for model, dfm in frontier_wide.items():
        ax.plot(-dfm["maxdd"], dfm["ret"], marker="o", label=model)
    ax.set_title(f"CAGR vs |MaxDD| {title_suffix}")
    ax.set_xlabel("|MaxDD|")
    ax.set_ylabel("CAGR")
    ax.grid(True, alpha=0.3)

    ax = axes[0, 2]
    for model, dfm in frontier_wide.items():
        ax.plot(dfm["ulcer"], dfm["ret"], marker="o", label=model)
    ax.set_title(f"CAGR vs Ulcer Index {title_suffix}")
    ax.set_xlabel("Ulcer Index")
    ax.set_ylabel("CAGR")
    ax.grid(True, alpha=0.3)

    ax = axes[1, 0]
    for model, dfm in frontier_wide.items():
        ax.plot(dfm.index, dfm["sharpe"], marker="o", label=model)
    ax.set_title(f"Sharpe vs tau {title_suffix}")
    ax.set_xlabel("tau")
    ax.set_ylabel("Sharpe")
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=8)

    ax = axes[1, 1]
    for model, dfm in frontier_wide.items():
        ax.plot(dfm.index, dfm["martin"], marker="o", label=model)
    ax.set_title(f"Martin Ratio vs tau {title_suffix}")
    ax.set_xlabel("tau")
    ax.set_ylabel("Martin Ratio")
    ax.grid(True, alpha=0.3)

    ax = axes[1, 2]
    for model, dfm in frontier_wide.items():
        ax.plot(dfm.index, dfm["equity_weight"], marker="o", label=model)
    ax.set_title(f"Equity weight vs tau {title_suffix}")
    ax.set_xlabel("tau")
    ax.set_ylabel(f"{EQUITY} weight")
    ax.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()


def best_points_by_metric(tbl, metrics=("sharpe", "calmar", "martin", "ret"), exclude_spy_only=False):
    df = tbl.copy()
    if exclude_spy_only:
        df = df[~df["is_spy_only"]].copy()

    rows = []
    for metric in metrics:
        valid = df.dropna(subset=[metric])
        if valid.empty:
            continue
        idx = valid.groupby("model")[metric].idxmax()
        tmp = valid.loc[idx, [
            "model", "tau", "ret", "vol", "sharpe", "maxdd",
            "calmar", "ulcer", "martin", "n_assets", "equity_weight",
            "is_spy_only", "is_core_satellite",
        ]]
        tmp = tmp.assign(best_by=metric)
        rows.append(tmp)

    return pd.concat(rows, axis=0).sort_values(["best_by", "model"]) if rows else pd.DataFrame()


def weights_panel(weights_by_model_tau, model_name):
    rows = []
    for (m, tau), w in weights_by_model_tau.items():
        if m != model_name:
            continue
        ws = pd.Series(w).astype(float).copy()
        ws.name = float(tau)
        rows.append(ws)
    if not rows:
        return pd.DataFrame()
    out = pd.DataFrame(rows)
    out.index.name = "tau"
    return out.sort_index()


def make_train_obj_for_kappa(kappa, base_obj=None):
    if base_obj is None:
        base_obj = train_soft_frob

    soft_k = compute_soft_frobenius_cov(
        R_train, window=WINDOW, kappa=float(kappa), clip_z=3.0,
        shrink_delta="auto", label=f"soft-frob-kappa-{kappa:g}"
    )
    # Keep growth estimates fixed to isolate covariance-conditioning effect.
    soft_k.update({
        "Sigma_base": base_obj["Sigma_base"],
        "Sigma_growth": base_obj["Sigma_growth"],
        "mu_mean": base_obj["mu_mean"],
        "boot_means": base_obj.get("boot_means", None),
        "kappa": float(kappa),
    })
    return soft_k


def kappa_sensitivity_summary(kappa_grid, tau_grid, R_eval):
    rows = []
    all_tables = {}

    for kappa in kappa_grid:
        obj_k = make_train_obj_for_kappa(kappa)
        tbl_k, _ = build_ablation_frontiers(
            obj_k, tau_grid, R_eval,
            w_core_min=W_CORE_MIN, k=K_SOCP, tickers=TICKERS
        )
        tbl_k["kappa"] = float(kappa)
        all_tables[float(kappa)] = tbl_k

        # Summarise best non-degenerate Greedy and Stress-MV performance.
        for model in ["Greedy Soft-Frob", "Stress-MV Soft-Frob"]:
            sub = tbl_k[(tbl_k["model"] == model) & (~tbl_k["is_spy_only"])].dropna(subset=["martin"])
            if sub.empty:
                rows.append({
                    "kappa": float(kappa),
                    "model": model,
                    "n_eff": obj_k["n_eff"],
                    "shrink_delta": obj_k["shrink_delta"],
                    "best_tau": np.nan,
                    "best_martin": np.nan,
                    "best_sharpe": np.nan,
                    "best_ret": np.nan,
                    "best_maxdd": np.nan,
                    "best_ulcer": np.nan,
                    "best_n_assets": np.nan,
                    "best_equity_weight": np.nan,
                })
                continue

            best = sub.loc[sub["martin"].idxmax()]
            rows.append({
                "kappa": float(kappa),
                "model": model,
                "n_eff": obj_k["n_eff"],
                "shrink_delta": obj_k["shrink_delta"],
                "best_tau": best["tau"],
                "best_martin": best["martin"],
                "best_sharpe": best["sharpe"],
                "best_ret": best["ret"],
                "best_maxdd": best["maxdd"],
                "best_ulcer": best["ulcer"],
                "best_n_assets": best["n_assets"],
                "best_equity_weight": best["equity_weight"],
            })

    return pd.DataFrame(rows), all_tables


def select_kappa_tau_paper_safe(selection_tbl,
                                metric="val_martin",
                                min_neff=None,
                                rel_tol=0.02,
                                prefer="higher_neff"):
    """
    Select (kappa, tau) from a validation table.

    The rule is deliberately conservative: among candidates close to the best
    validation score, prefer the one based on more effective observations.
    """
    if min_neff is None:
        min_neff = max(10 * len(TICKERS), 200)

    tbl = selection_tbl.copy().dropna(subset=[metric, "n_eff", "kappa", "tau"])
    tbl = tbl[tbl["n_eff"] >= min_neff].copy()

    if tbl.empty:
        raise ValueError(
            f"No candidates satisfy min_neff={min_neff}. "
            "Lower min_neff or expand the training window."
        )

    best_val = tbl[metric].max()
    cutoff = best_val * (1 - rel_tol) if best_val >= 0 else best_val * (1 + rel_tol)
    near_best = tbl[tbl[metric] >= cutoff].copy()

    if prefer == "higher_neff":
        selected = near_best.sort_values(["n_eff", metric], ascending=[False, False]).iloc[0]
    elif prefer == "lower_kappa":
        selected = near_best.sort_values(["kappa", metric], ascending=[True, False]).iloc[0]
    elif prefer == "best_metric":
        selected = tbl.sort_values(metric, ascending=False).iloc[0]
    else:
        raise ValueError("prefer must be one of {'higher_neff', 'lower_kappa', 'best_metric'}")

    return selected, tbl.sort_values(metric, ascending=False), near_best.sort_values(metric, ascending=False)


def rebuild_final_results_for_selected_params(kappa_selected, tau_selected, label="paper-safe"):
    """Re-estimate on train+validation and evaluate once on final test."""
    required = ["R_trainval", "logR_trainval", "R_final"]
    missing = [x for x in required if x not in globals()]
    if missing:
        raise RuntimeError(f"Missing required objects from Section 15: {missing}")

    Sigma_base_tv = compute_cov_lw(R_trainval, label=f"trainval-full-lw-{label}")
    boot_tv = stationary_bootstrap_means(logR_trainval.values, BLOCK_SIZE, N_BOOT)
    mu_tv = pd.Series(boot_tv.mean(0) * 252, index=R_trainval.columns)
    Sigma_growth_tv = pd.DataFrame(
        np.cov(boot_tv * 252, rowvar=False),
        index=R_trainval.columns, columns=R_trainval.columns
    )

    soft_tv = compute_soft_frobenius_cov(
        R_trainval, window=WINDOW, kappa=float(kappa_selected), clip_z=3.0,
        shrink_delta="auto", label=f"trainval-soft-frob-{label}"
    )

    obj_tv = {
        **soft_tv,
        "Sigma_base": Sigma_base_tv,
        "Sigma_growth": Sigma_growth_tv,
        "mu_mean": mu_tv,
        "kappa": float(kappa_selected),
    }

    tbl, weights = build_ablation_frontiers(
        obj_tv, [float(tau_selected)], R_final,
        w_core_min=W_CORE_MIN, k=K_SOCP, tickers=TICKERS
    )
    return tbl, weights, obj_tv
```

### 2.3 Transaction costs, walk-forward backtests, plotting and inference

```python
def make_static_target_schedule(index, weights, rebal_freq=21):
    """Target-weight schedule for a fixed allocation rebalanced every rebal_freq days."""
    w = pd.Series(weights).reindex(TICKERS).fillna(0.0).astype(float)
    if w.sum() > 0:
        w = w / w.sum()
    rebal_dates = index[::rebal_freq]
    return pd.DataFrame(
        np.tile(w.values, (len(rebal_dates), 1)),
        index=rebal_dates,
        columns=TICKERS,
    )


def backtest_target_weight_schedule_net(R_eval, target_schedule, half_spread,
                                        include_initial_cost=False):
    """
    Backtest a target-weight schedule with daily returns and transaction costs.

    Rebalancing cost at date t:
        sum_i |target_i - pre_trade_weight_i| * half_spread_i(t)

    Gross and net returns should be computed with this same engine; gross simply
    uses a zero cost panel. This avoids mixing daily-rebalanced static returns
    with drifted target-weight returns.
    """
    R_eval = R_eval.copy()
    half_spread = (
        half_spread.reindex(R_eval.index).ffill().bfill()
        .reindex(columns=R_eval.columns).fillna(0.0)
    )

    target_schedule = target_schedule.reindex(columns=R_eval.columns).sort_index()
    target_schedule = target_schedule.loc[target_schedule.index.intersection(R_eval.index)]

    current_w = np.zeros(len(R_eval.columns), dtype=float)
    gross_rets, net_rets, costs, turnovers, weights_used = [], [], [], [], []

    for dt, r in R_eval.iterrows():
        if dt in target_schedule.index:
            target = target_schedule.loc[dt].values.astype(float)
            target = np.maximum(target, 0.0)
            if target.sum() > 0:
                target = target / target.sum()

            if (current_w.sum() == 0) and (not include_initial_cost):
                turnover = 0.0
                tc = 0.0
            else:
                trade = target - current_w
                turnover = float(np.abs(trade).sum())
                hs = half_spread.loc[dt].values.astype(float)
                tc = float(np.nansum(np.abs(trade) * hs))
            current_w = target.copy()
        else:
            turnover = 0.0
            tc = 0.0

        r_vec = r.values.astype(float)
        gross = float(current_w @ r_vec)
        net = float((1.0 - tc) * (1.0 + gross) - 1.0)

        gross_rets.append(gross)
        net_rets.append(net)
        costs.append(tc)
        turnovers.append(turnover)
        weights_used.append(current_w.copy())

        # Drift weights after returns.
        denom = 1.0 + gross
        if denom > 0:
            current_w = current_w * (1.0 + r_vec) / denom
            current_w = np.maximum(current_w, 0.0)
            if current_w.sum() > 0:
                current_w = current_w / current_w.sum()

    return {
        "gross": pd.Series(gross_rets, index=R_eval.index, name="gross"),
        "net": pd.Series(net_rets, index=R_eval.index, name="net"),
        "cost": pd.Series(costs, index=R_eval.index, name="tc"),
        "turnover": pd.Series(turnovers, index=R_eval.index, name="turnover"),
        "weights_daily": pd.DataFrame(weights_used, index=R_eval.index, columns=R_eval.columns),
    }


def _extract_ohlc_panel(raw, field, tickers=TICKERS):
    """Robust extraction from yfinance multi-index OHLC output."""
    if isinstance(raw.columns, pd.MultiIndex):
        if field in raw.columns.get_level_values(0):
            out = raw[field]
        elif field in raw.columns.get_level_values(1):
            out = raw.xs(field, level=1, axis=1)
        else:
            raise KeyError(f"Could not find field {field} in OHLC data.")
    else:
        out = raw[[field]]
    return out.reindex(columns=tickers).astype(float)


def _clean_spread_panel(full_spread, window_fill=21, q_clip=(0.01, 0.99)):
    """
    Clean proportional full-spread estimates without imposing exogenous flat costs.
    - Missing values: rolling median -> expanding median -> cross-sectional median.
    - Blow-ups: empirical per-asset clipping.
    """
    x = full_spread.replace([np.inf, -np.inf], np.nan).astype(float)
    x = x.where(x >= 0)

    roll = x.rolling(window_fill, min_periods=3).median()
    x = x.combine_first(roll)

    expm = x.expanding(min_periods=5).median()
    x = x.combine_first(expm)

    xs_med = x.median(axis=1)
    x = x.T.fillna(xs_med).T
    x = x.ffill().bfill()

    lo = x.quantile(q_clip[0])
    hi = x.quantile(q_clip[1])
    x = x.clip(lower=lo, upper=hi, axis=1)
    return x


def corwin_schultz_full_spread(high, low, window_fill=21):
    """Corwin-Schultz high-low proportional full bid-ask spread estimate."""
    H = high.astype(float).replace(0, np.nan)
    L = low.astype(float).replace(0, np.nan)

    log_hl = np.log(H / L).replace([np.inf, -np.inf], np.nan)
    beta = log_hl.pow(2) + log_hl.shift(1).pow(2)

    H2 = pd.concat([H, H.shift(1)], keys=["t", "t-1"]).groupby(level=1).max()
    L2 = pd.concat([L, L.shift(1)], keys=["t", "t-1"]).groupby(level=1).min()
    gamma = np.log(H2 / L2).replace([np.inf, -np.inf], np.nan).pow(2)

    denom = 3 - 2 * np.sqrt(2)
    alpha = (np.sqrt(2 * beta) - np.sqrt(beta)) / denom - np.sqrt(gamma / denom)
    alpha = alpha.clip(lower=0)

    full_spread = 2 * (np.exp(alpha) - 1) / (1 + np.exp(alpha))
    return _clean_spread_panel(full_spread, window_fill=window_fill)


def abdi_ranaldo_full_spread(close, high, low, window_fill=21):
    """
    Abdi-Ranaldo-style close/high/low proportional full-spread proxy.

    Implementation uses log close c_t and log midrange eta_t=(h_t+l_t)/2.
    The raw spread proxy is based on the serial product of close-midrange deviations.
    We shift the resulting estimate by one day before use to avoid using t+1 information
    for same-day trading-cost accounting.
    """
    C = close.astype(float).replace(0, np.nan)
    H = high.astype(float).replace(0, np.nan)
    L = low.astype(float).replace(0, np.nan)

    c = np.log(C).replace([np.inf, -np.inf], np.nan)
    h = np.log(H).replace([np.inf, -np.inf], np.nan)
    l = np.log(L).replace([np.inf, -np.inf], np.nan)
    eta = 0.5 * (h + l)

    # AR-type raw estimator. The next-day eta creates a two-day estimate; shift later.
    raw = (c - eta) * (c - eta.shift(-1))
    raw = raw.clip(lower=0)
    s_log = 2.0 * np.sqrt(raw)

    full_spread = 2 * (np.exp(s_log) - 1) / (1 + np.exp(s_log))
    full_spread = full_spread.shift(1)  # no-lookahead when used as cost at date t
    return _clean_spread_panel(full_spread, window_fill=window_fill)


def load_ohlc_panels(tickers=TICKERS, start=START, end=END):
    raw = yf.download(tickers, start=start, end=end, auto_adjust=True, progress=False)
    return {
        "Open":  _extract_ohlc_panel(raw, "Open", tickers),
        "High":  _extract_ohlc_panel(raw, "High", tickers),
        "Low":   _extract_ohlc_panel(raw, "Low", tickers),
        "Close": _extract_ohlc_panel(raw, "Close", tickers),
    }


def data_driven_positive_floor(half_spread, window=252, q=0.10, min_periods=20):
    """
    Rolling/expanding floor using only positive historical estimates.
    This prevents zero-cost collapse while avoiding exogenous flat bps assumptions.
    """
    x = half_spread.astype(float).replace([np.inf, -np.inf], np.nan)
    pos = x.where(x > 0)

    # Use only past observations for rolling/expanding floors.
    pos_lag = pos.shift(1)
    roll_floor = pos_lag.rolling(window, min_periods=min_periods).quantile(q)
    exp_floor = pos_lag.expanding(min_periods=min_periods).quantile(q)
    floor = roll_floor.combine_first(exp_floor)

    # Cross-sectional same-date positive median as a data-driven fallback.
    xs_floor = pos.median(axis=1)
    floor = floor.T.fillna(xs_floor).T

    # Asset-level positive median as final data-driven fallback.
    asset_floor = pos.median(axis=0)
    floor = floor.fillna(asset_floor)

    # If an asset has no positive observations at all, leave floor at zero.
    floor = floor.fillna(0.0)
    return floor


def apply_data_driven_floor(half_spread, window=252, q=0.10):
    floor = data_driven_positive_floor(half_spread, window=window, q=q)
    return half_spread.combine(floor, np.maximum)


def stress_liquidity_multiplier(half_spread, stress_signal=None, strength=0.50, clip_z=3.0):
    """
    Regime-dependent transaction-cost stress: costs rise with positive Frobenius stress intensity.
    The multiplier is dimensionless; it does not impose a flat bps cost.
    """
    if stress_signal is None:
        if "frob_d" in globals():
            stress_signal = frob_d
        else:
            return half_spread.copy()

    s = pd.Series(stress_signal).reindex(half_spread.index).ffill().bfill().astype(float)

    # Expanding standardization using past observations where possible.
    mu = s.shift(1).expanding(min_periods=60).mean()
    sd = s.shift(1).expanding(min_periods=60).std(ddof=0).replace(0, np.nan)
    z = ((s - mu) / sd).replace([np.inf, -np.inf], np.nan).fillna(0.0).clip(-clip_z, clip_z)
    z_pos = z.clip(lower=0)

    mult = 1.0 + strength * z_pos
    return half_spread.mul(mult, axis=0)


def load_transaction_cost_models(tickers=TICKERS, start=START, end=END, align_index=None):
    """Build a panel of one-way cost estimates for transaction-cost robustness checks."""
    if align_index is None:
        align_index = R.index

    ohlc = load_ohlc_panels(tickers, start, end)
    cs_full = corwin_schultz_full_spread(ohlc["High"], ohlc["Low"])
    ar_full = abdi_ranaldo_full_spread(ohlc["Close"], ohlc["High"], ohlc["Low"])

    cs = (0.5 * cs_full).reindex(align_index).ffill().bfill().reindex(columns=tickers)
    ar = (0.5 * ar_full).reindex(align_index).ffill().bfill().reindex(columns=tickers)

    # Composites. Median is robust; max is conservative.
    med = pd.concat({"CS": cs, "AR": ar}, axis=1).groupby(level=1, axis=1).median().reindex(columns=tickers)
    mx = pd.concat({"CS": cs, "AR": ar}, axis=1).groupby(level=1, axis=1).max().reindex(columns=tickers)

    cs_floor = apply_data_driven_floor(cs, window=252, q=0.10)
    ar_floor = apply_data_driven_floor(ar, window=252, q=0.10)
    med_floor = apply_data_driven_floor(med, window=252, q=0.10)

    stress_med_floor = stress_liquidity_multiplier(med_floor, stress_signal=frob_d if "frob_d" in globals() else None,
                                                   strength=0.50, clip_z=3.0)

    models = {
        "CS_pure": cs,
        "AR_pure": ar,
        "CS_AR_median": med,
        "CS_AR_max": mx,
        "CS_floor_data_driven": cs_floor,
        "AR_floor_data_driven": ar_floor,
        "CS_AR_median_floor": med_floor,
        "CS_AR_median_floor_stress": stress_med_floor,
    }

    # Final alignment and cleaning.
    models = {k: v.reindex(align_index).ffill().bfill().reindex(columns=tickers).fillna(0.0)
              for k, v in models.items()}
    return models


def summarize_cost_model(cost_panel):
    x = (cost_panel * 1e4).stack().replace([np.inf, -np.inf], np.nan).dropna()
    return pd.Series({
        "mean_bps": x.mean(),
        "median_bps": x.median(),
        "p75_bps": x.quantile(0.75),
        "p90_bps": x.quantile(0.90),
        "p95_bps": x.quantile(0.95),
        "max_bps": x.max(),
        "zero_share": float((cost_panel <= 0).stack().mean()),
    })


def evaluate_static_models_under_cost_models(R_eval, final_weights, final_tbl, cost_models, rebal_freq=21):
    """Evaluate final static model weights under each transaction-cost model."""
    rows = []
    ret_store = {}

    if not {"final_weights", "final_tbl"}.issubset(globals()) and (final_weights is None or final_tbl is None):
        return pd.DataFrame(), {}

    model_list = list(final_tbl["model"].unique())
    for cost_name, half_spread in cost_models.items():
        for model in model_list:
            Wm = weights_panel(final_weights, model)
            if Wm.empty:
                continue
            w_model = Wm.iloc[0].reindex(TICKERS).fillna(0.0)
            sched = make_static_target_schedule(R_eval.index, w_model, rebal_freq=rebal_freq)
            bt = backtest_target_weight_schedule_net(R_eval, sched, half_spread, include_initial_cost=False)

            gross_perf = extended_perf_summary(bt["gross"])
            net_perf = extended_perf_summary(bt["net"])

            row = {"cost_model": cost_name, "model": model}
            row.update({f"gross_{k}": v for k, v in gross_perf.items()})
            row.update({f"net_{k}": v for k, v in net_perf.items()})
            row["avg_turnover_ann"] = float(bt["turnover"].mean() * 252)
            row["avg_tc_ann"] = float(bt["cost"].mean() * 252)
            row["total_tc"] = float(bt["cost"].sum())
            rows.append(row)

            ret_store[(cost_name, model, "gross")] = bt["gross"]
            ret_store[(cost_name, model, "net")] = bt["net"]

    out = pd.DataFrame(rows)
    if not out.empty:
        out = out.set_index(["cost_model", "model"]).sort_index()
    return out, ret_store


def _build_growth_estimates(logR_train, n_boot=None):
    if n_boot is None:
        n_boot = WF_N_BOOT

    boot = stationary_bootstrap_means(logR_train.values, BLOCK_SIZE, n_boot)

    mu = pd.Series(
        boot.mean(0) * 252,
        index=logR_train.columns
    )

    Sigma_g = pd.DataFrame(
        np.cov(boot * 252, rowvar=False),
        index=logR_train.columns,
        columns=logR_train.columns
    )

    return mu, Sigma_g


def _tau_grid_from_growth(mu, Sigma_g, n_grid=9):
    eq_sd = float(np.sqrt(max(Sigma_g.loc[EQUITY, EQUITY], 0)))
    eq_robust = float(mu[EQUITY] - K_SOCP * eq_sd)

    tau_lo = float(mu.min() - 0.06)
    tau_hi = float(eq_robust * 0.995)

    return np.linspace(tau_lo, tau_hi, n_grid)


def _select_greedy_soft_frob_on_validation(
    R_train_sel,
    R_val_sel,
    logR_train_sel,
    kappa_grid=None,
    tau_grid=None,
    min_neff=None,
    n_boot=None,
    metric="martin",
    rel_tol=None,
):
    """Select (kappa, tau) on validation using the paper-safe near-best / high-N_eff rule."""

    if kappa_grid is None:
        kappa_grid = WF_KAPPA_GRID
    if min_neff is None:
        min_neff = WF_MIN_NEFF
    if n_boot is None:
        n_boot = WF_N_BOOT
    if rel_tol is None:
        rel_tol = WF_REL_TOL

    Sigma_base = compute_cov_lw(R_train_sel, label="wf-train-base")

    mu, Sigma_g = _build_growth_estimates(
        logR_train_sel,
        n_boot=n_boot
    )

    if tau_grid is None:
        tau_grid = _tau_grid_from_growth(mu, Sigma_g, n_grid=9)

    rows = []

    for kappa in kappa_grid:
        soft = compute_soft_frobenius_cov(
            R_train_sel,
            window=WINDOW,
            kappa=float(kappa),
            clip_z=3.0,
            shrink_delta="auto",
            label=f"wf-kappa-{kappa:g}",
        )

        if soft["n_eff"] < min_neff:
            continue

        obj = {
            **soft,
            "Sigma_base": Sigma_base,
            "Sigma_growth": Sigma_g,
            "mu_mean": mu,
            "kappa": float(kappa),
        }

        for tau in tau_grid:
            res = solve_greedy_point(
                obj["Sigma_stress"],
                obj["Sigma_growth"],
                obj["mu_mean"],
                tau=float(tau),
                w_core_min=W_CORE_MIN,
                k=K_SOCP,
                tickers=TICKERS,
            )

            w = res["weights"]
            flags = _weight_flags(w)

            if w.isna().any() or flags["is_spy_only"]:
                continue

            ret_val = portfolio_returns_static(R_val_sel, w)
            perf = extended_perf_summary(ret_val)

            perf.update({
                "kappa": float(kappa),
                "tau": float(tau),
                "n_eff": soft["n_eff"],
                "shrink_delta": soft["shrink_delta"],
                **flags,
            })

            rows.append(perf)

    val_tbl = pd.DataFrame(rows)

    if val_tbl.empty:
        return None, None, None

    best_val = val_tbl[metric].max()

    if best_val >= 0:
        cutoff = best_val * (1.0 - rel_tol)
    else:
        cutoff = best_val * (1.0 + rel_tol)

    near = val_tbl[val_tbl[metric] >= cutoff].copy()

    selected = near.sort_values(
        ["n_eff", metric],
        ascending=[False, False]
    ).iloc[0]

    return selected, val_tbl, tau_grid


def _backtest_continuous_schedule(
    R_all,
    schedule,
    half_spread=None,
    include_initial_cost=False,
):
    """Compute gross and net returns using the same target-weight drift/rebalance engine."""

    schedule = schedule.sort_index()

    if schedule.empty:
        return None, None

    start = schedule.index.min()

    # Include all dates from first schedule date to the end of the available OOS panel.
    # Since the walk-forward loop creates schedules until the final available fold, this
    # keeps the last selected target allocation active through the final OOS date.
    R_eval = R_all.loc[start:]

    if R_eval.empty:
        return None, None

    zero_cost = pd.DataFrame(
        0.0,
        index=R_eval.index,
        columns=R_eval.columns
    )

    gross_bt = backtest_target_weight_schedule_net(
        R_eval,
        schedule,
        zero_cost,
        include_initial_cost=include_initial_cost,
    )

    if half_spread is None:
        net_bt = None
    else:
        net_bt = backtest_target_weight_schedule_net(
            R_eval,
            schedule,
            half_spread,
            include_initial_cost=include_initial_cost,
        )

    return gross_bt, net_bt


def run_walk_forward_oos(
    R_all,
    logR_all,
    kappa_grid=None,
    train_days=756,
    val_days=252,
    test_step=21,
    n_boot=None,
    min_neff=None,
    rel_tol=None,
    reestimate_mode="rolling",
    include_transaction_costs=True,
    half_spread=None,
):
    """
    Rolling/expanding walk-forward OOS.

    Returns
    -------
    fold_tbl : DataFrame
        Per-fold diagnostics and selected parameters.
    gross_panel : dict[str, Series]
        Continuous gross OOS returns, computed by target-weight schedule engine.
    net_panel : dict[str, Series]
        Continuous net OOS returns after transaction costs.
    weight_schedules : dict[str, DataFrame]
        Fold-by-fold target weights.
    impl_tbl : DataFrame
        Continuous implementation diagnostics: turnover and cost.
    """

    if kappa_grid is None:
        kappa_grid = WF_KAPPA_GRID
    if min_neff is None:
        min_neff = WF_MIN_NEFF
    if n_boot is None:
        n_boot = WF_N_BOOT
    if rel_tol is None:
        rel_tol = WF_REL_TOL

    rows = []
    schedules = {}

    start = train_days + val_days
    fold = 0

    for test_start in range(start, len(R_all), test_step):
        test_end = min(test_start + test_step, len(R_all))

        if test_end <= test_start:
            break

        if reestimate_mode == "expanding":
            train_start = 0
        else:
            train_start = max(0, test_start - val_days - train_days)

        train_end = test_start - val_days
        val_start = train_end
        val_end = test_start

        if train_end - train_start < train_days // 2:
            continue
        if val_end - val_start < max(63, val_days // 3):
            continue

        fold += 1

        R_tr = R_all.iloc[train_start:train_end]
        R_va = R_all.iloc[val_start:val_end]
        R_te = R_all.iloc[test_start:test_end]
        logR_tr = logR_all.iloc[train_start:train_end]

        selected, val_tbl, tau_grid = _select_greedy_soft_frob_on_validation(
            R_tr,
            R_va,
            logR_tr,
            kappa_grid=kappa_grid,
            tau_grid=None,
            min_neff=min_neff,
            n_boot=n_boot,
            metric="martin",
            rel_tol=rel_tol,
        )

        if selected is None:
            continue

        kappa_star = float(selected["kappa"])
        tau_star = float(selected["tau"])

        R_tv = R_all.iloc[train_start:val_end]
        logR_tv = logR_all.iloc[train_start:val_end]

        Sigma_base_tv = compute_cov_lw(
            R_tv,
            label=f"wf-fold-{fold}-base-tv"
        )

        mu_tv, Sigma_g_tv = _build_growth_estimates(
            logR_tv,
            n_boot=n_boot
        )

        soft_tv = compute_soft_frobenius_cov(
            R_tv,
            window=WINDOW,
            kappa=kappa_star,
            clip_z=3.0,
            shrink_delta="auto",
            label=f"wf-fold-{fold}-soft-tv",
        )

        obj_tv = {
            **soft_tv,
            "Sigma_base": Sigma_base_tv,
            "Sigma_growth": Sigma_g_tv,
            "mu_mean": mu_tv,
            "kappa": kappa_star,
        }

        fold_tbl, fold_weights = build_ablation_frontiers(
            obj_tv,
            [tau_star],
            R_te,
            w_core_min=W_CORE_MIN,
            k=K_SOCP,
            tickers=TICKERS,
        )

        for model in fold_tbl["model"].unique():
            Wm = weights_panel(fold_weights, model)

            if Wm.empty:
                continue

            w_model = Wm.iloc[0].reindex(TICKERS).fillna(0.0)

            sched = pd.DataFrame(
                [w_model.values],
                index=[R_te.index[0]],
                columns=TICKERS,
            )

            schedules.setdefault(model, []).append(sched)

            flags = _weight_flags(w_model)

            rows.append({
                "fold": fold,
                "model": model,
                "train_start": R_tr.index[0],
                "train_end": R_tr.index[-1],
                "val_start": R_va.index[0],
                "val_end": R_va.index[-1],
                "test_start": R_te.index[0],
                "test_end": R_te.index[-1],
                "kappa": kappa_star,
                "tau": tau_star,
                "n_eff_tv": soft_tv["n_eff"],
                "shrink_delta_tv": soft_tv["shrink_delta"],
                **flags,
            })

    fold_diag = pd.DataFrame(rows)

    weight_schedules = {
        model: pd.concat(parts).sort_index()
        for model, parts in schedules.items()
    }

    gross_panel = {}
    net_panel = {}
    impl_rows = []

    for model, sched in weight_schedules.items():
        gross_bt, net_bt = _backtest_continuous_schedule(
            R_all,
            sched,
            half_spread=half_spread if include_transaction_costs else None,
            include_initial_cost=False,
        )

        if gross_bt is None:
            continue

        gross_panel[model] = gross_bt["gross"]

        row = {
            "model": model,
            "n_rebalances": int(len(sched)),
            "avg_turnover_ann_gross_engine": float(gross_bt["turnover"].mean() * 252),
        }

        if net_bt is not None:
            net_panel[model] = net_bt["net"]
            row.update({
                "avg_turnover_ann": float(net_bt["turnover"].mean() * 252),
                "avg_tc_ann": float(net_bt["cost"].mean() * 252),
                "total_tc": float(net_bt["cost"].sum()),
            })
        else:
            row.update({
                "avg_turnover_ann": np.nan,
                "avg_tc_ann": np.nan,
                "total_tc": np.nan,
            })

        impl_rows.append(row)

    if impl_rows:
        impl_tbl = pd.DataFrame(impl_rows).set_index("model")
    else:
        impl_tbl = pd.DataFrame()

    return fold_diag, gross_panel, net_panel, weight_schedules, impl_tbl


def _summarize_return_dict(ret_dict, version, protocol):
    rows = []
    for model, ret in ret_dict.items():
        row = {"protocol": protocol, "model": model, "version": version}
        row.update(extended_perf_summary(ret))
        rows.append(row)
    return rows


def run_walk_forward_sensitivity(R_all, logR_all, protocols=None,
                                 kappa_grid=None,
                                 n_boot=None,
                                 min_neff=None,
                                 half_spread=None,
                                 include_transaction_costs=True):
    """Run a compact sensitivity panel over walk-forward protocol choices."""
    if protocols is None:
        protocols = [
            {"name": "rolling_3y_1y_1m", "train_days": 756,  "val_days": 252, "test_step": 21, "mode": "rolling"},
            {"name": "rolling_5y_1y_1m", "train_days": 1260, "val_days": 252, "test_step": 21, "mode": "rolling"},
            {"name": "expanding_1y_1m",   "train_days": 756,  "val_days": 252, "test_step": 21, "mode": "expanding"},
        ]

    if kappa_grid is None:
        kappa_grid = WF_KAPPA_GRID
    if n_boot is None:
        n_boot = WF_N_BOOT
    if min_neff is None:
        min_neff = WF_MIN_NEFF

    all_fold_tables, all_gross, all_net, all_impl = {}, {}, {}, {}
    summary_rows, selection_rows, impl_rows = [], [], []

    for cfg in protocols:
        name = cfg["name"]
        print(f"\n=== Walk-forward sensitivity: {name} ===")
        folds, gross, net, schedules, impl = run_walk_forward_oos(
            R_all, logR_all,
            kappa_grid=kappa_grid,
            train_days=cfg["train_days"],
            val_days=cfg["val_days"],
            test_step=cfg["test_step"],
            n_boot=n_boot,
            min_neff=min_neff,
            reestimate_mode=cfg["mode"],
            include_transaction_costs=include_transaction_costs and (half_spread is not None),
            half_spread=half_spread,
        )

        all_fold_tables[name] = folds
        all_gross[name] = gross
        all_net[name] = net
        all_impl[name] = impl

        summary_rows.extend(_summarize_return_dict(gross, "gross", name))
        if net:
            summary_rows.extend(_summarize_return_dict(net, "net", name))

        if not folds.empty:
            sel = folds.groupby("model").agg(
                n_folds=("fold", "nunique"),
                median_kappa=("kappa", "median"),
                mean_kappa=("kappa", "mean"),
                median_tau=("tau", "median"),
                mean_neff=("n_eff_tv", "mean"),
                median_neff=("n_eff_tv", "median"),
                mean_n_assets=("n_assets", "mean"),
            ).reset_index()
            sel.insert(0, "protocol", name)
            selection_rows.append(sel)

        if not impl.empty:
            tmp = impl.reset_index().rename(columns={"index": "model"})
            tmp.insert(0, "protocol", name)
            impl_rows.append(tmp)

    summary = pd.DataFrame(summary_rows)
    if not summary.empty:
        summary = summary.set_index(["protocol", "model", "version"]).sort_index()

    selection_summary = pd.concat(selection_rows, ignore_index=True) if selection_rows else pd.DataFrame()
    if not selection_summary.empty:
        selection_summary = selection_summary.set_index(["protocol", "model"]).sort_index()

    implementation_summary = pd.concat(impl_rows, ignore_index=True) if impl_rows else pd.DataFrame()
    if not implementation_summary.empty:
        implementation_summary = implementation_summary.set_index(["protocol", "model"]).sort_index()

    return {
        "folds": all_fold_tables,
        "gross": all_gross,
        "net": all_net,
        "impl": all_impl,
        "summary": summary,
        "selection_summary": selection_summary,
        "implementation_summary": implementation_summary,
    }


def _get_wf_return_dict(protocol: str = "rolling_3y_1y_1m", version: str = "net"):
    """
    Retrieve walk-forward return dictionary.
    protocol options typically include:
      - 'rolling_3y_1y_1m'
      - 'rolling_5y_1y_1m'
      - 'expanding_1y_1m'
    """
    # Prefer protocol-sensitivity object if available.
    if "wf_sensitivity" in globals():
        source = wf_sensitivity.get(version, {})
        if protocol in source:
            return source[protocol]

    # Fall back to canonical walk-forward.
    if protocol == "rolling_3y_1y_1m":
        if version == "net" and "wf_net" in globals():
            return wf_net
        if version == "gross" and "wf_gross" in globals():
            return wf_gross

    raise ValueError(
        f"Could not find {version} returns for protocol={protocol}. "
        "Run the canonical walk-forward and/or protocol-sensitivity blocks first."
    )


def plot_wf_wealth_and_drawdown(
    ret_dict,
    title: str,
    out_path: str,
    model_order=None,
):
    if model_order is None:
        model_order = MODEL_ORDER

    cols = [m for m in model_order if m in ret_dict]
    if not cols:
        raise ValueError("None of the requested models are present in ret_dict.")

    ret_panel = pd.concat({m: ret_dict[m] for m in cols}, axis=1).dropna()
    wealth = (1.0 + ret_panel).cumprod()
    wealth = wealth / wealth.iloc[0]

    drawdown = wealth / wealth.cummax() - 1.0

    fig, axes = plt.subplots(
        2, 1,
        figsize=(9.0, 6.5),
        sharex=True,
        gridspec_kw={"height_ratios": [2.0, 1.0]},
    )

    for m in cols:
        axes[0].plot(wealth.index, wealth[m], label=m, linewidth=1.6)
    axes[0].set_title(title)
    axes[0].set_ylabel("Cumulative wealth")
    axes[0].legend(loc="best", fontsize=8)
    axes[0].grid(True, alpha=0.3)

    for m in cols:
        axes[1].plot(drawdown.index, 100.0 * drawdown[m], label=m, linewidth=1.3)
    axes[1].set_ylabel("Drawdown (%)")
    axes[1].set_xlabel("Date")
    axes[1].grid(True, alpha=0.3)

    fig.tight_layout()
    fig.savefig(out_path, dpi=300, bbox_inches="tight")
    plt.show()

    return ret_panel, wealth, drawdown


def subperiod_performance(return_panel, periods=None):
    """
    return_panel : dict[str, Series]
        Model name -> daily return series.
    periods : dict[str, tuple[str,str]]
        Label -> (start_date, end_date).
    """
    if periods is None:
        periods = {
            "COVID shock/rebound": ("2020-02-01", "2020-12-31"),
            "Rates/bond bear market": ("2022-01-01", "2022-12-31"),
            "Recovery / disinflation": ("2023-01-01", "2023-12-31"),
            "Recent test window": ("2024-01-01", "2025-12-31"),
            "Full available panel": (None, None),
        }

    rows = []
    for model, ret in return_panel.items():
        r = ret.dropna()
        for label, (start, end) in periods.items():
            sub = r.copy()
            if start is not None:
                sub = sub.loc[sub.index >= pd.Timestamp(start)]
            if end is not None:
                sub = sub.loc[sub.index <= pd.Timestamp(end)]
            if len(sub) < 30:
                continue
            perf = extended_perf_summary(sub)
            perf.update({"model": model, "period": label, "n_days": len(sub)})
            rows.append(perf)

    return pd.DataFrame(rows).set_index(["period", "model"]).sort_index()


def circular_block_bootstrap_indices(n, block_len=21, n_boot=1000, random_state=42):
    rng = np.random.default_rng(random_state)
    out = np.empty((n_boot, n), dtype=int)

    for b in range(n_boot):
        idx = []
        while len(idx) < n:
            start = rng.integers(0, n)
            block = (start + np.arange(block_len)) % n
            idx.extend(block.tolist())
        out[b, :] = np.array(idx[:n])
    return out


def metric_value(ret, metric):
    p = extended_perf_summary(pd.Series(ret))
    return p[metric]


def paired_bootstrap_metric_diff(ret_a, ret_b,
                                 metrics=("sharpe", "martin", "ulcer", "maxdd", "ret"),
                                 block_len=21, n_boot=1000, random_state=42):
    """Bootstrap CI for metric(a)-metric(b), using paired circular-block resampling."""
    df = pd.concat([ret_a.rename("a"), ret_b.rename("b")], axis=1).dropna()
    a = df["a"].values
    b = df["b"].values
    n = len(df)

    idx_mat = circular_block_bootstrap_indices(n, block_len=block_len, n_boot=n_boot, random_state=random_state)

    rows = []
    for metric in metrics:
        obs = metric_value(a, metric) - metric_value(b, metric)
        boot_vals = np.empty(n_boot)

        for j in range(n_boot):
            idx = idx_mat[j]
            boot_vals[j] = metric_value(a[idx], metric) - metric_value(b[idx], metric)

        ci_lo, ci_hi = np.nanpercentile(boot_vals, [2.5, 97.5])
        p_gt0 = float(np.mean(boot_vals > 0))
        p_lt0 = float(np.mean(boot_vals < 0))

        rows.append({
            "metric": metric,
            "obs_diff": obs,
            "ci_2.5": ci_lo,
            "ci_97.5": ci_hi,
            "p(diff>0)": p_gt0,
            "p(diff<0)": p_lt0,
            "n_obs": n,
            "block_len": block_len,
        })

    return pd.DataFrame(rows).set_index("metric")


def run_pairwise_inference(return_panel, pairs=None, block_len=21, n_boot=1000):
    if pairs is None:
        pairs = [
            ("Greedy Soft-Frob", "Markowitz Base-Cov"),
            ("Stress-MV Soft-Frob", "Markowitz Base-Cov"),
            ("Greedy Soft-Frob", "Stress-MV Soft-Frob"),
            ("Greedy Soft-Frob", "Greedy Base-Cov"),
        ]

    tables = {}
    for a, b in pairs:
        if a not in return_panel or b not in return_panel:
            continue
        tbl = paired_bootstrap_metric_diff(
            return_panel[a], return_panel[b],
            block_len=block_len, n_boot=n_boot, random_state=42
        )
        tables[(a, b)] = tbl
    return tables
```

## 2. Stress Indicators

```python
# ── 2a. Equicorrelation via MIR (Main Identifying Restriction) ──────────
# ρ(t) = (σ²_portfolio - Σ w_i² σ_i²) / Σ_{i≠j} w_i w_j σ_i σ_j
# When ρ rises, all pairwise correlations converge → diversification fails

print('Computing equicorrelation...')
equicorr = rolling_equicorr(R, W_lag, WINDOW)
print(f'  range: [{equicorr.min():.3f}, {equicorr.max():.3f}]')
```

```python
# ── 2b. dCor-Pearson gap  ──────────────────────────────────────────────
# Gap(t) = w' dCor(t) w  −  w' |Pearson(t)| w
# Interpretation: "excess non-linear dependence" beyond linear dependence in modulus.
# NOTE: dCor is more expensive than Pearson; for larger universes consider:
#   - working at asset-class level (small N),
#   - sampling pairs, or
#   - computing dCor only on a subset of risky assets.

print('Computing dCor-|Pearson| gap (can take a few minutes)...')
gap = rolling_dcor_gap(R, W_lag, WINDOW)
print(f'  range: [{gap.min():.4f}, {gap.max():.4f}]')
```

```python
# ── 2c. Frobenius distance from long-run correlation ──────────────────
# Measures how far the rolling correlation matrix has drifted from a
# long-run 'normal' structure. Large distance = structural shift.
#
# IMPORTANT: for out-of-sample work, compute rho_lt on TRAIN only (or expanding up to t-1)
# to avoid look-ahead.

print('Computing Frobenius distance...')
frob_d = rolling_frobenius(R, WINDOW)
print(f'  range: [{frob_d.min():.3f}, {frob_d.max():.3f}]')
```

## 5. Log-Growth Proxy & Stationary Bootstrap

```python
# Politis-Romano stationary bootstrap for robust growth estimates.
# mu_robust = mean_boot - k * std_boot  (conservative lower bound)
# Used in SOCP growth constraint to ensure robustness to estimation noise.

K_BOOTSTRAP = 1.0    # confidence level
N_BOOT      = 1000
BLOCK_SIZE  = 22     # 1 trading month blocks

print(f'Running stationary bootstrap (B={N_BOOT}, block={BLOCK_SIZE})...')
boot_means = stationary_bootstrap_means(logR.values, BLOCK_SIZE, N_BOOT)

mu_boot_mean = boot_means.mean(axis=0)
mu_boot_std  = boot_means.std(axis=0)
mu_robust_annual = pd.Series((mu_boot_mean - K_BOOTSTRAP * mu_boot_std) * 252, index=R.columns)
mu_mean_annual   = pd.Series(mu_boot_mean * 252, index=R.columns)
# Growth uncertainty covariance (annualised) from bootstrap mean distribution
boot_means_annual = boot_means * 252
Sigma_growth = pd.DataFrame(np.cov(boot_means_annual, rowvar=False), index=R.columns, columns=R.columns)
print('\nSigma_growth (annual) diag (std of annual log-growth mean estimates):')
print(np.sqrt(np.diag(Sigma_growth)).round(4))

print(pd.DataFrame({'Point estimate': mu_mean_annual,
                    'Robust (mean-k*sd)': mu_robust_annual}).round(4))

# Plot bootstrap distributions
fig, axes = plt.subplots(2, 3, figsize=(15, 7))
for i, (ax, t) in enumerate(zip(axes.flat, TICKERS)):
    ax.hist(boot_means[:, i] * 252, bins=40, color='steelblue', alpha=0.7, ec='white')
    ax.axvline(mu_mean_annual.iloc[i],   color='red',    lw=2, label='Mean')
    ax.axvline(mu_robust_annual.iloc[i], color='orange', lw=2, ls='--', label='Robust')
    ax.set_title(t, fontsize=11); ax.legend(fontsize=7); ax.grid(True, alpha=0.3)
plt.suptitle('Bootstrap distribution of annualised log-growth', fontsize=13)
plt.tight_layout(); plt.show()
```

## 7. Greedy Outer Loop (Core-Satellite Construction)

```python
# Economic logic:
# - Equity (EQUITY) is the growth engine: we enforce a minimum allocation w_core_min.
# - We add satellites greedily: at each step pick the one that most reduces
#   stress variance while satisfying the robust growth constraint.
# - The greedy loop stops when no further satellite improves stress resilience.

# ── Demo at tau = 0 ────────────────────────────────────────────────────
W_CORE_MIN = 0.50
K_SOCP     = 1.00

print('='*65)
print(f'Greedy portfolio  tau=0  w_core_min={W_CORE_MIN}')
print('='*65)
res0 = greedy_core_satellite(TICKERS, Sigma_stress, Sigma_growth, mu_mean_annual,
                              w_core_min=W_CORE_MIN, tau=0.0, k=K_SOCP, verbose=True)
print('\nOptimal weights:')
print(res0['weights'].round(4).to_string())
print(f"\nAnnualised stress vol : {res0['stress_vol_annual']:.4f}")
print(f"Mean growth (ann.)     : {res0['mean_growth_annual']:+.4f}")
print(f"Robust growth (ann.)   : {res0['robust_growth_annual']:+.4f}")
print(f"Selected assets        : {res0['selected']}")
```

## 8. τ-Grid Frontier

```python
# Sweep tau from conservative (low robust growth requirement) to aggressive.
# Each point on the frontier is a full greedy solve.

# Feasible tau range should be compared to robust growth, not raw mean.
eq = EQUITY
eq_sd = float(np.sqrt(max(Sigma_growth.loc[eq, eq], 0)))
eq_robust = float(mu_mean_annual[eq] - K_SOCP * eq_sd)

tau_lo   = float(mu_mean_annual.min() - 0.10)   # very permissive
tau_hi   = float(eq_robust * 0.98)              # near equity robust growth
TAU_GRID = np.linspace(tau_lo, tau_hi, 18)

frontier = []
print('Computing tau-grid frontier...')
for tau in TAU_GRID:
    res = greedy_core_satellite(TICKERS, Sigma_stress, Sigma_growth, mu_mean_annual,
                                 w_core_min=W_CORE_MIN, tau=tau, k=K_SOCP, verbose=False)
    frontier.append(res)
    print(f"  tau={tau:+.3f}  sv={res['stress_vol_annual']:.4f}  "
          f"meanG={res['mean_growth_annual']:+.4f}  robustG={res['robust_growth_annual']:+.4f}  "
          f"n_assets={len(res['selected'])}")

taus    = np.array([f['tau'] for f in frontier])
s_vols  = np.array([f['stress_vol_annual'] for f in frontier])
robG    = np.array([f['robust_growth_annual'] for f in frontier])

weight_df = pd.DataFrame([f['weights'].values for f in frontier],
                          columns=TICKERS, index=np.round(taus, 3))
weight_df.index.name = 'tau'

# ── Plots ─────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(18, 4))

axes[0].plot(taus, s_vols, marker='o', color='royalblue')
axes[0].set_title('Stress-Vol vs tau')
axes[0].set_xlabel('tau (annual robust growth threshold)')
axes[0].set_ylabel('Stress Vol (ann.)')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(s_vols, robG, s=60, c='crimson')
axes[1].set_title('Risk-Return Frontier (stress-aware)')
axes[1].set_xlabel('Stress Vol (ann.)')
axes[1].set_ylabel('Robust Egrowth (ann.)')
axes[1].grid(True, alpha=0.3)

weight_df.plot(kind='bar', stacked=True, ax=axes[2], width=0.9)
axes[2].set_title('Portfolio Composition along Frontier')
axes[2].set_xlabel('tau')
axes[2].set_ylabel('Weight')
axes[2].legend(loc='upper left', bbox_to_anchor=(1.02, 1.0), frameon=False)

plt.tight_layout()
plt.show()
```

```python
# Holdout train/test split (parameters estimated on train, then applied OOS).
# NOTE: this is still a holdout setup, not a full rolling walk-forward.

TRAIN_FRAC = 0.70
TAU_BT     = 0.0  # can be changed after looking at the train frontier

split_idx = int(TRAIN_FRAC * len(R))
R_train   = R.iloc[:split_idx]
R_test    = R.iloc[split_idx:]
logR_tr   = logR.iloc[:split_idx]

print(f'Train: {R_train.index[0].date()} - {R_train.index[-1].date()} ({len(R_train)} days)')
print(f'Test : {R_test.index[0].date()} - {R_test.index[-1].date()} ({len(R_test)} days)')

print('\nRe-estimating training objects (primary = equicorr)...')
train_eq = estimate_train_objects(R_train, logR_tr, indicator='equicorr',
                                  window=WINDOW, percentile=PERCENTILE,
                                  block_size=BLOCK_SIZE, n_boot=N_BOOT)

Sigma_stress_tr = train_eq['Sigma_stress']
Sigma_base_tr   = train_eq['Sigma_base']
mu_mean_tr      = train_eq['mu_mean']
Sigma_growth_tr = train_eq['Sigma_growth']

print('Mean mu (train, ann.):', mu_mean_tr.round(4).to_dict())
print('Growth sigma (train, ann.) diag:', np.sqrt(np.diag(Sigma_growth_tr)).round(4))
```

```python
# ── Holdout backtest (fixed-parameter OOS) ─────────────────────────────
# Note: this is a "train -> test holdout" with parameters estimated on train
# and then rebalanced in test using the SAME (fixed) Sigma/mu.
# A true walk-forward version would re-estimate Sigma/mu on a rolling/expanding window.

print('Running holdout backtest...')
strat_ret = holdout_rebalance_backtest(R_test, TICKERS, Sigma_stress_tr, Sigma_growth_tr,
                                       mu_mean_tr, w_core_min=W_CORE_MIN,
                                       tau=TAU_BT, k=K_SOCP, rebal_freq=21)

# Benchmarks
# Equal weight
ew_w = np.ones(len(TICKERS)) / len(TICKERS)
ew_ret = pd.Series(R_test.values @ ew_w, index=R_test.index)

# Equity-only
eq_idx = TICKERS.index(EQUITY)
eq_w = np.zeros(len(TICKERS)); eq_w[eq_idx] = 1.0
eq_ret = pd.Series(R_test.values @ eq_w, index=R_test.index)

# Markowitz min-variance (full-sample cov on train), long-only, sum=1
# NOTE: this is a baseline, not stress-aware.
w = cp.Variable(len(TICKERS))
Ss = Sigma_base_tr.loc[TICKERS, TICKERS].values
prob = cp.Problem(cp.Minimize(cp.quad_form(w, Ss + 1e-10*np.eye(len(TICKERS)))),
                  [cp.sum(w)==1, w>=0])
try:
    prob.solve(solver=cp.OSQP, verbose=False)
except Exception:
    prob.solve(solver=cp.SCS, verbose=False)
mkw_w = w.value if w.value is not None else ew_w
mkw_w = np.maximum(mkw_w, 0); mkw_w = mkw_w / mkw_w.sum()
mkw_ret = pd.Series(R_test.values @ mkw_w, index=R_test.index)

print('Backtest done.')
```

## 11. V1.1 — Soft Frobenius Stress Weighting + Efficient Frontier Comparison

This section implements the next step of the framework:

1. Use **Frobenius distance** as a continuous regime intensity variable, rather than a hard percentile classifier.
2. Estimate a **soft stress-weighted covariance matrix**:
   \[
   \omega_t(\kappa) \propto \exp(\kappa z_t), \qquad
   z_t = \frac{D_t - \bar D}{s_D}
   \]
   where \(D_t\) is the Frobenius distance from the long-run correlation structure.
3. Build the **Greedy SAA frontier** by sweeping \(\tau\).
4. Build a **Markowitz robust-growth frontier** on the same \(\tau\)-grid.
5. Compare the frontiers on several OOS metrics: CAGR, volatility, Sharpe, MaxDD, Calmar, Ulcer Index, and Martin Ratio.

The three state variables remain separate in the earlier section; here we promote **soft Frobenius** as the current main specification for the efficient-frontier test.


## 14. Sensitivity to \(\kappa\)

\(\kappa\) controls how aggressively the Frobenius regime intensity weights the covariance estimator toward stress observations.

This block does not re-bootstrap growth for each \(\kappa\); it keeps \(\mu_b^{log}\) and \(\Sigma_b\) fixed from the main train object to isolate the effect of the stress covariance only.


## 15. Paper-grade train / validation / test scaffold

This block selects \((\kappa, \tau)\) on a validation window, then evaluates once on a final test window.

Current split:
- training selection window: first 60%
- validation window: next 20%
- final test window: last 20%

Selection metric: Martin Ratio on validation, excluding SPY-only endpoints.


```python

# ── Train / validation / test selection scaffold ──────────────────────────
# This is the first paper-grade selection protocol.
# It selects kappa and tau on validation, excluding SPY-only endpoints,
# then evaluates the selected specification once on the final test window.

RUN_TVT_VALIDATION = True

if RUN_TVT_VALIDATION:
    tr_end = int(0.60 * len(R))
    va_end = int(0.80 * len(R))

    R_sel_train = R.iloc[:tr_end]
    R_val       = R.iloc[tr_end:va_end]
    R_final     = R.iloc[va_end:]

    logR_sel_train = logR.iloc[:tr_end]
    R_trainval     = R.iloc[:va_end]
    logR_trainval  = logR.iloc[:va_end]

    print(f"Selection train: {R_sel_train.index[0].date()} - {R_sel_train.index[-1].date()} ({len(R_sel_train)})")
    print(f"Validation     : {R_val.index[0].date()} - {R_val.index[-1].date()} ({len(R_val)})")
    print(f"Final test     : {R_final.index[0].date()} - {R_final.index[-1].date()} ({len(R_final)})")

    # Growth estimates computed once on selection train to avoid bootstrap noise across kappa.
    Sigma_base_sel = compute_cov_lw(R_sel_train, label="sel-train-full-lw")
    boot_sel = stationary_bootstrap_means(logR_sel_train.values, BLOCK_SIZE, N_BOOT)
    mu_sel = pd.Series(boot_sel.mean(0) * 252, index=R_sel_train.columns)
    Sigma_growth_sel = pd.DataFrame(
        np.cov(boot_sel * 252, rowvar=False),
        index=R_sel_train.columns, columns=R_sel_train.columns
    )

    eq_sd_sel = float(np.sqrt(max(Sigma_growth_sel.loc[EQUITY, EQUITY], 0)))
    eq_robust_sel = float(mu_sel[EQUITY] - K_SOCP * eq_sd_sel)
    tau_sel_lo = float(mu_sel.min() - 0.06)
    tau_sel_hi = float(eq_robust_sel * 0.995)
    TAU_GRID_SELECTION = np.linspace(tau_sel_lo, tau_sel_hi, 15)

    selection_rows = []

    for kappa in KAPPA_GRID:
        soft_sel = compute_soft_frobenius_cov(
            R_sel_train, window=WINDOW, kappa=float(kappa), clip_z=3.0,
            shrink_delta="auto", label=f"sel-soft-frob-kappa-{kappa:g}"
        )
        obj_sel = {
            **soft_sel,
            "Sigma_base": Sigma_base_sel,
            "Sigma_growth": Sigma_growth_sel,
            "mu_mean": mu_sel,
            "kappa": float(kappa),
        }

        tbl_val, _ = build_ablation_frontiers(
            obj_sel, TAU_GRID_SELECTION, R_val,
            w_core_min=W_CORE_MIN, k=K_SOCP, tickers=TICKERS
        )

        # Select only the full framework, excluding the SPY-only boundary.
        sub = tbl_val[
            (tbl_val["model"] == "Greedy Soft-Frob") &
            (~tbl_val["is_spy_only"])
        ].dropna(subset=["martin"])

        if sub.empty:
            continue

        best = sub.loc[sub["martin"].idxmax()]
        selection_rows.append({
            "kappa": float(kappa),
            "n_eff": soft_sel["n_eff"],
            "shrink_delta": soft_sel["shrink_delta"],
            "tau": float(best["tau"]),
            "val_ret": best["ret"],
            "val_vol": best["vol"],
            "val_sharpe": best["sharpe"],
            "val_maxdd": best["maxdd"],
            "val_ulcer": best["ulcer"],
            "val_martin": best["martin"],
            "val_n_assets": best["n_assets"],
            "val_equity_weight": best["equity_weight"],
        })

    selection_tbl = pd.DataFrame(selection_rows).sort_values("val_martin", ascending=False)
    print("\nValidation selection table (Greedy Soft-Frob, SPY-only endpoint excluded):")
    display(selection_tbl.round(4))

    selected = selection_tbl.iloc[0]
    kappa_star = float(selected["kappa"])
    tau_star = float(selected["tau"])

    print(f"\nSelected parameters: kappa*={kappa_star:g}, tau*={tau_star:+.4f}")

    # Re-estimate on train+validation, then evaluate once on final test.
    Sigma_base_tv = compute_cov_lw(R_trainval, label="trainval-full-lw")
    boot_tv = stationary_bootstrap_means(logR_trainval.values, BLOCK_SIZE, N_BOOT)
    mu_tv = pd.Series(boot_tv.mean(0) * 252, index=R_trainval.columns)
    Sigma_growth_tv = pd.DataFrame(
        np.cov(boot_tv * 252, rowvar=False),
        index=R_trainval.columns, columns=R_trainval.columns
    )
    soft_tv = compute_soft_frobenius_cov(
        R_trainval, window=WINDOW, kappa=kappa_star, clip_z=3.0,
        shrink_delta="auto", label="trainval-soft-frob-selected"
    )

    obj_tv = {
        **soft_tv,
        "Sigma_base": Sigma_base_tv,
        "Sigma_growth": Sigma_growth_tv,
        "mu_mean": mu_tv,
        "kappa": kappa_star,
    }

    final_tbl, final_weights = build_ablation_frontiers(
        obj_tv, [tau_star], R_final,
        w_core_min=W_CORE_MIN, k=K_SOCP, tickers=TICKERS
    )

    print("\nFinal test results at validation-selected (kappa*, tau*):")
    display(final_tbl.round(4))

    print("\nFinal test weights:")
    for model in final_tbl["model"].unique():
        Wm = weights_panel(final_weights, model)
        print(f"\n{model}")
        display(Wm.round(3))
else:
    print("RUN_TVT_VALIDATION=False. Set it to True to run train/validation/test selection.")
```

## 21. Inference: paired circular-block bootstrap on performance differentials

We estimate confidence intervals for performance **differences** between models using a paired block bootstrap of daily returns.  
This preserves time-series dependence better than iid resampling and focuses on model-vs-model deltas.

```python
# ── Inference for paired performance differentials ─────────────────────────

RUN_INFERENCE = True
INFERENCE_SOURCE = "auto"  # options: auto, final_net, wf_net, final_gross, wf_gross

if RUN_INFERENCE:
    if INFERENCE_SOURCE in ("auto", "final_net") and "final_model_returns_net" in globals() and len(final_model_returns_net) > 0:
        inference_returns = final_model_returns_net
        inf_label = "net final-test returns under main TC model"
    elif INFERENCE_SOURCE in ("auto", "wf_net") and "wf_net" in globals() and len(wf_net) > 0:
        inference_returns = wf_net
        inf_label = "net walk-forward returns — continuous schedule engine"
    elif INFERENCE_SOURCE in ("auto", "final_gross") and "final_model_returns_gross" in globals() and len(final_model_returns_gross) > 0:
        inference_returns = final_model_returns_gross
        inf_label = "gross final-test returns"
    elif INFERENCE_SOURCE in ("auto", "wf_gross") and "wf_gross" in globals() and len(wf_gross) > 0:
        inference_returns = wf_gross
        inf_label = "gross walk-forward returns — continuous schedule engine"
    else:
        inference_returns = None
        inf_label = None

    if inference_returns is not None:
        print(f"Running paired circular-block bootstrap inference on {inf_label}...")
        inference_tables = run_pairwise_inference(
            inference_returns,
            block_len=21,
            n_boot=1000,
        )

        for (a, b), tbl in inference_tables.items():
            print(f"\n{a} minus {b}:")
            display(tbl.round(4))
    else:
        print("No return panel found for inference. Run Section 17/18 first.")
else:
    print("RUN_INFERENCE=False. Set to True to run bootstrap inference.")
```
